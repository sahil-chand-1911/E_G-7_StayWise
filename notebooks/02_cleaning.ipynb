# 02: Data Cleaning and Feature Engineering

```python
import pandas as pd
import numpy as np

listings = pd.read_pickle('../listings_raw.pkl')
past_rates = pd.read_pickle('../past_rates_raw.pkl')

print(f"Loaded {len(listings)} listings.")
```

## 1. Handling Missing Values

```python
cat_cols = ['city', 'listing_type', 'room_type', 'cancellation_policy']
for col in cat_cols:
    if col in listings.columns:
        listings[col] = listings[col].fillna('Unknown')

listings['ttm_revenue'] = pd.to_numeric(listings.get('ttm_revenue', 0)).fillna(0)
listings['ttm_occupancy'] = pd.to_numeric(listings.get('ttm_occupancy', 0)).fillna(0)

listings = listings[listings['ttm_revenue'] > 0].copy()
print(f"Listings with revenue: {len(listings)}")
```

## 2. Processing Amenities

```python
listings['amenities'] = listings.get('amenities', '').fillna('')
listings['amenity_list'] = listings['amenities'].apply(lambda x: [a.strip().lower() for a in str(x).split(',') if a.strip()])
listings['amenity_count'] = listings['amenity_list'].apply(len)

print("Processed amenities.")
```

## 3. Save Cleaned Data

```python
def ensure_stable_types(df):
    for col in df.columns:
        if pd.api.types.is_string_dtype(df[col]):
            df[col] = df[col].astype(object)
    return df

listings = ensure_stable_types(listings)
past_rates = ensure_stable_types(past_rates)

listings.to_pickle('../listings_cleaned.pkl', protocol=4)
past_rates.to_pickle('../past_rates_cleaned.pkl', protocol=4)
print("Cleaned data saved as Pickle")
```
