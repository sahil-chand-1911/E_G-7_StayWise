# 01: Data Loading

```python
import pandas as pd
import numpy as np
import os
```

## Load Datasets

```python
LISTINGS_PATH = '../data/raw/listings.csv'
RATES_PATH = '../data/raw/past_rates.csv'

try:
    listings = pd.read_csv(LISTINGS_PATH)
except Exception as e:
    listings = pd.read_csv(LISTINGS_PATH, on_bad_lines='skip')

past_rates = pd.read_csv(RATES_PATH)

def ensure_stable_types(df):
    for col in df.columns:
        if pd.api.types.is_string_dtype(df[col]):
            df[col] = df[col].astype(object)
    return df

listings = ensure_stable_types(listings)
past_rates = ensure_stable_types(past_rates)

print(f"Listings shape: {listings.shape}")
print(f"Past Rates shape: {past_rates.shape}")
```

## Save Intermediate Data

```python
listings.to_pickle('../listings_raw.pkl', protocol=4)
past_rates.to_pickle('../past_rates_raw.pkl', protocol=4)
print("Raw data saved to Pickle files")
```
